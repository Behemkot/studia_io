```python
import numpy as np
import pandas as pd
import os
from itertools import combinations
```

```python
df = pd.read_csv("./dane/SzerPerm.csv", sep=";")
```

```python
df = df.drop(["Lp.", "Nazwa cżęści"], axis=1)
```

```python
def wartosc_uporzadkowania(df):
    obrobione = df.copy()
    obrobione["Odlewanie"] = df["Odlewanie"].cumsum()
    obrobione.iloc[0] = df.iloc[0].cumsum()
    
    for i in range(1, len(obrobione)):
        for j in range(1, len(obrobione.iloc[0])):
            max_val = max(int(obrobione.iloc[[i-1], [j]].values),
                             int(obrobione.iloc[[i], [j-1]].values))
            obrobione.iloc[[i], [j]] = max_val + df.iloc[[i], [j]]
    wartosc = int(obrobione.iloc[[-1], [-1]].values)
    return obrobione, wartosc
```

```python
obrobione, wartosc = wartosc_uporzadkowania(df)
wartosc
```
