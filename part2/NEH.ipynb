```python
import pandas as pd
import numpy as np
import os
from matplotlib.pyplot import plot as plt

data_path = "dane/"
```

```python
df = pd.read_excel(os.path.join(data_path, "dane1.xlsx"))
```

```python
df["Wi"] = df.sum(axis=1) - df["Zadanie"]
df = df.sort_values(by=["Wi"], ascending=False)
df = df.drop(["Wi"], axis=1)
```

```python
def get_value_of_order(df):
    if len(df) <= 0:
        return 0
    elif len(df) == 1:
        return int(df.sum(axis=1))

    obrobione = df.copy()
    obrobione["M1"] = df["M1"].cumsum() # TODO
    obrobione.iloc[1] = df.iloc[1].cumsum()
    
    for i in range(2, len(obrobione)):
        for j in range(2, len(obrobione.iloc[1])):
            max_val = max(int(obrobione.iloc[[i-1], [j]].values),
                             int(obrobione.iloc[[i], [j-1]].values))
            obrobione.iloc[[i], [j]] = max_val + df.iloc[[i], [j]]
    wartosc = int(obrobione.iloc[[-1], [-1]].values)
    return wartosc

def swap(a, b, df = df):
    df.iloc[a], df.iloc[b] = df.iloc[b].copy(), df.iloc[a].copy()
```

```python
def NEH(df = df):
    '''
    Wykomentować jakie parametry przyjmuje
    i reszte
    '''
    df.index = range(len(df))
    order = pd.DataFrame()
    value_of_order = np.inf

    for index, row in df.iterrows():
        row = pd.DataFrame(row).T
        order = order.append(row, ignore_index = True)
        value_of_order = get_value_of_order(order)

        new_order = order.copy()
        length = index + 1
        print(f"Kombinacje dla {index + 1} elementów")
        for idx in range(length):
            new_value = get_value_of_order(new_order)

            if new_value < value_of_order:
                print(f"{value_of_order} -> {new_value}")
                order = new_order.copy()
                value_of_order = new_value

            if idx != length - 1:
                a, b = length - idx - 2, length - idx - 1
                new_order.iloc[a], new_order.iloc[b] = new_order.iloc[b].copy(), new_order.iloc[a].copy()

    return order, value_of_order
```
